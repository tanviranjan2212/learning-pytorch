### Playing around with Seaborn

Playing around with plots from the [Seaborn Gallery](https://seaborn.pydata.org/examples/index.html)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pdb
from tqdm import tqdm
```

```python
sns.set(style="ticks")
df = sns.load_dataset("anscombe")

sns.lmplot(x='x',y='y',col='dataset',data=df,hue='dataset',col_wrap=4);
plt.show()
```

```python
fmri = sns.load_dataset("fmri"); fmri.head()
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='region',style='event'); sns.despine()
```

```python
dots = sns.load_dataset("dots"); dots.head()
sns.relplot(x='time',y='firing_rate',data=dots,hue='coherence',size='choice',
            col='align',kind='line',facet_kws=dict(sharex=False));
plt.show()
```

```python
penguins = sns.load_dataset("penguins"); penguins.head()
f,ax=plt.subplots(1,2,figsize=(15,5))
sns.catplot(data=penguins, kind='bar',x='species', y='body_mass_g',hue='sex',ci='sd',ax=ax[0]);
sns.despine()

sns.jointplot(data=penguins,x='bill_length_mm',y='bill_depth_mm',hue='species',kind='kde',ax=ax[1])
plt.show()
```

```python
tips = sns.load_dataset("tips"); tips.head()
sns.violinplot(data=tips,x='day',y='total_bill',hue='smoker',split=True,inner='quart');
sns.despine()
```

```python
planets = sns.load_dataset("planets"); planets.head()
f,ax=plt.subplots()
sns.boxplot(data=planets,x='distance',y='method',ax=ax,whis=[0,100]); sns.despine();
ax.set_xscale('log');
sns.stripplot(data=planets,x='distance',y='method',ax=ax,color='black')
```
