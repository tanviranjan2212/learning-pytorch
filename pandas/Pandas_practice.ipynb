#### Combining dataframes
Join exercises [here ](https://www.w3resource.com/python-exercises/pandas/joining-and-merging/index.php)

1. Write a Pandas program to join the two given dataframes along rows and assign all data

```python
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})

all_student = pd.concat([student_data1, student_data2]);
all_student
```

2. Write a Pandas program to join the two given dataframes along columns and assign all

```python
all_student_col = pd.concat([student_data1, student_data2], axis=1);
all_student_col
```

3. Write a Pandas program to append rows to an existing DataFrame and display the combined data

```python
s6 = pd.Series(['S6', 'Scarlette Fisher', 205], index=['student_id', 'name', 'marks'])
student_11 = student_data1.append(s6, ignore_index=True);
student_11
```

5. Write a Pandas program to join the two given dataframes along rows and merge with another dataframe along the common column id

```python
exam_data = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
        'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]})

pd.merge(all_student,exam_data, on='student_id')
```

6. Write a Pandas program to join the two dataframes using the common column of both dataframes

7. Write a Pandas program to join the two dataframes with matching records from both sides where available

8. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.

```python
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})

pd.merge(data1, data2, on=['key1','key2'],how='left')
```

15. Write a Pandas program to Combine two DataFrame objects by filling null values in one DataFrame with non-null values from other DataFrame.

8. **Write a Pandas program to create a Pivot table and find manager wise, salesman wise total sale and also display the sum of all sale amount at the bottom**

Pretty cool functionality of using margins=True to print all sum

```python
df.pivot_table(index=['Manager','SalesMan'],values=['Sale_amt','Units'],aggfunc=[np.sum], margins=True)
```

9. Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise, sales man wise where Manager = "Douglas"
